Encontrar, para um único país (ex Italia), regras usando o algoritmo apriori, como visto no exemplo no jupyter. 
Utilizar medidas de suporte, confiança e lift 
Testar alguns valores 
Definir possíveis melhores valores para tais medidas utilizadas

```python
#conda install mlxtend
#conda install mlxtend --channel conda-forge

import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('Online Retail.xlsx')
print len (df)
df.head()
```

```python
df_italy = df.loc[df["Country"] == "Italy"] 

print len(df_italy)

df_italy.head()
```

```python
columns_to_filter = ['UnitPrice',"StockCode",'CustomerID']
df_italy = df_italy.loc[:, columns_to_filter]
df_italy['UnitPrice'] = df_italy['UnitPrice'].astype('str')
df_italy['CustomerID'] = df_italy['CustomerID'].astype('str')
df_italy.head()
```

```python
df1 = pd.get_dummies(df_italy)
df1.head()
```

```python
#para ter regras com mais de um item no itemsets o support não pode ser maior que 0.01 
#pelo print abaixo dá para ver que quando muda para 0.02 restam poucos itemsets com mais de um item
frequent_items = apriori(df1, min_support=0.01, use_colnames=True)
frequent_items.sort_values('support', ascending= True).head(50)
```

```python
#se definir threshold maior que 0.5 restam poucas regras a serem avaliadas
rules = association_rules(frequent_items, metric='confidence', min_threshold=0.5)
rules
```

```python
#10 primeiras regras com maior suporte
rules.sort_values('support', ascending= False).head(10)
```

```python
#10 primeiras regras com maior confianca
rules.sort_values('confidence', ascending= False).head(10)
```

```python
#10 primeiras regras com maior lift
rules.sort_values('lift', ascending= False).head(10)
```

```python
#selecionando regras com lift maior que 2, confianca maior que 0.6 e suporte maior que 0.2
rules[(rules['lift'] >= 2) & (rules['confidence'] >= 0.6) & (rules['support'] >= 0.01 )]
```

```python
#encontrar regras com lift maior que 1
rules = association_rules(frequent_items, metric='lift', min_threshold=1)
#type(rules)
#rules.shape
rules
```

```python
#10 primeiras regras com maior suporte
rules.sort_values('support', ascending= False).head(10)
```

```python
#10 primeiras regras com maior confianca
rules.sort_values('confidence', ascending= False).head(10)
```

```python
#10 primeiras regras com maior lift
rules.sort_values('lift', ascending= False).head(10)
```

```python
#selecionando regras com lift maior que 2, confianca maior que 0.6 e suporte maior que 0.01
rules[(rules['lift'] >= 2) & (rules['confidence'] >= 0.6) & (rules['support'] >= 0.01 )]
```

a metrica confidence apresenta os melhores resultados
apesar do support ser baixo o valor conviction dos resultados é alto, inclusive alcançando o valor inf
