```python
#conda install mlxtend
#conda install mlxtend --channel conda-forge

import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('Online Retail.xlsx')
print len (df)
df.head()
```

```python
df_italy = df.loc[df["Country"] == "Italy"] 

print len(df_italy)

#df_italy.head()
```

```python
columns_to_filter = ["CustomerID","StockCode",'UnitPrice']
df_italy = df_italy.loc[:, columns_to_filter]
df_italy['UnitPrice'] = df_italy['UnitPrice'].astype('str')
df_italy['CustomerID'] = df_italy['CustomerID'].astype('str')
```

```python
df1 = pd.get_dummies(df_italy)
df1.head()
```

```python
#encontrando regras com no mínimo 5% de suporte
frequent_items = apriori(df1, min_support=0.01, use_colnames=True)
frequent_items
```

```python
#encontrar regras com lift maior que 1
#rules = association_rules(frequent_items, metric='lift', min_threshold=1)
rules = association_rules(frequent_items, metric='confidence', min_threshold=0.01)
#rules = association_rules(frequent_items, metric='confidence')
type(rules)
rules.shape
rules
```

```python
#10 primeiras regras com maior suporte
rules.sort_values('support', ascending= False).head(10)
```

```python
#10 primeiras regras com maior confianca
rules.sort_values('confidence', ascending= False).head(10)
```

```python
#10 primeiras regras com maior lift
rules.sort_values('lift', ascending= False).head(10)
```

```python
#selecionando regras com lift maior que 2, confianca maior que 0.6 e suporte maior que 0.2
rules[(rules['lift'] >= 2) & (rules['confidence'] >= 0.01) & (rules['support'] >= 0.01 )]
```
