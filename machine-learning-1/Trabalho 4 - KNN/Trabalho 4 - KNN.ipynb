#### Define o método para calcular a acurácia do modelo

```python
def getAccuracy(testset, predictions):
    correct = 0
    for id_test, test in testset.iterrows():
        if test[-1] == predictions.ix[id_test][0]:
            correct += 1
    return (correct / float(len(testset))) * 100.0

```

```python
scores = pd.DataFrame(columns=["n_of_k", 
                               "accuracy",
                              "%_of_sample"])

sample_splits = [0.15, 0.2, 0.3, 0.4]
k_max = 5

########################################
# Variando os dados de treino e teste
########################################

for i in sample_splits:
    print "Sampling by "+str(i)+"%"
    trainset, testset, = train_test_split(zoo, test_size=i)
    print len(trainset)
    print len(testset)

    ########################################
    # Para cada variação, executa de K a Max
    ########################################

    for k in range(1, k_max):
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(trainset.iloc[:, 1:], trainset.iloc[:, 17])
        predicted_knn = knn.predict(testset.iloc[:, 1:])
        accuracy = getAccuracy(testset, pd.DataFrame(predicted_knn, index=testset.index.values))
        #Add to score dataframe
        scores.loc[len(scores)] = [k, accuracy, i]
        print str(k) + "NN's : Acc " + str(accuracy) +" : Sample by "+str(i)+"% "
    
    print
```
